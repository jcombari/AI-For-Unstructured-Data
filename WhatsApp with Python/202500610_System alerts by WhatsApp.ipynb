```python
import pandas as pd
import matplotlib.pyplot as plt
import pywhatkit as kit
import datetime
import random

# Simulate a dataset of system logs
log_levels = ["INFO", "WARNING", "ERROR", "DEBUG"]
messages = [
    "System started",
    "User login successful",
    "Disk usage high",
    "Connection timeout",
    "ERROR: Unable to connect to database",
    "Process completed",
    "ERROR: Memory leak detected",
    "INFO: Health check passed"
]

# Generate sample log data
log_data = {
   "timestamp": pd.date_range(start="2025-01-01", periods=50, freq="h"),

    "log": [random.choice(messages) for _ in range(50)]
}

# Create DataFrame
logs = pd.DataFrame(log_data)

# Display the first few rows of the logs
print(logs.head())

# Detect if there are any ERROR logs
error_logs = logs[logs['log'].str.contains("ERROR")]

# Visualize the count of different log types
log_counts = logs['log'].apply(lambda x: "ERROR" if "ERROR" in x else "OTHER").value_counts()
log_counts.plot(kind='bar', color=["red", "green"])
plt.title("Log Type Distribution")
plt.xlabel("Log Type")
plt.ylabel("Count")
plt.show()

# If ERROR logs exist, send WhatsApp alert
if not error_logs.empty:
    now = datetime.datetime.now()
    send_time = now + datetime.timedelta(minutes=1)

    # Format message
    message = f"üö® ALERT: Found {len(error_logs)} ERROR logs in the system! Check immediately."

    # Send WhatsApp message (change phone number format with your number, e.g., '+34123456789')
    kit.sendwhatmsg("+34611105348", message, send_time.hour, send_time.minute)

    print("WhatsApp alert scheduled!")
else:
    print("No ERROR logs detected. No alert needed.")
```
