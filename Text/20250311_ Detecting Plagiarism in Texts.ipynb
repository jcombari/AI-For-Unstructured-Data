# 🐍 Python: Detectar Plagio en Textos con Pandas 📝

## ✨ ¿Es tu contenido realmente único?
Imagina que subes un artículo a la web y, semanas después, alguien lo publica con pequeñas modificaciones. ¿Cómo detectar rápidamente el plagio en bases de datos textuales?

## 👉 Solución: Usar **Levenshtein Distance** o **difflib** en Python para medir similitudes entre textos.

### 🔧 ¿Cómo funciona?
- Calculamos la distancia entre caracteres de dos textos.
- Cuanto menor la distancia, más similares son los textos.
- Convertimos la distancia en un puntaje de similitud.

### 🔎 ¿Por qué importa?
- Detecta contenido duplicado en grandes bases de datos.
- Ayuda a identificar fraudes en investigaciones académicas.
- Optimiza motores de búsqueda para evitar redundancias.

## ✨ Ejemplo real:
Usamos Pandas y la distancia de Levenshtein para medir la similitud entre textos con errores tipográficos. Así podrías identificar copias alteradas de un documento original.

### 🛠️ Impacto en el negocio:
- 📈 Empresas de contenido pueden proteger sus derechos de autor.
- 🌐 Motores de búsqueda pueden mejorar la relevancia de resultados.
- 🏫 Universidades pueden prevenir el plagio académico.

## 📊 Resumen del código:
- Comparamos textos con errores tipográficos.
- Medimos similitud usando Levenshtein Distance.
- Visualizamos resultados con un heatmap.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from Levenshtein import distance as levenshtein_distance

# Sample text pairs with varying levels of similarity
text_pairs = [
    ("machine learning", "machine learning"),  # Distance 0 (identical)
    ("machine learning", "machine earning"),   # Distance 2 (0-10 range)
    ("deep learning model", "deep larning modle"),  # Distance 5 (0-10 range)
    ("artificial intelligence", "artifical inteligence research"),  # Distance 12 (10-20 range)
    ("natural language processing", "natral lnguage procesing and analysis"),  # Distance 18 (10-20 range)
    ("computer vision applications", "compter vison apps and algorithms"),  # Distance 30 (20-40 range)
    ("big data and analytics in cloud computing", "b1g dat4 aNd an4lyt1cs in cl0ud c0mput1ng"),  # Distance 42 (20-40 range)
    ("generative adversarial networks for image generation", "g3ner4tiv3 adv3rsari4l n3tw0rks f0r 1m4g3 g3n3rati0n and synthesis"),  # Distance 65 (>60 range)
]

# Compute distances and similarity scores
data = []
for text1, text2 in text_pairs:
    dist = levenshtein_distance(text1, text2)
    similarity = 1 - (dist / max(len(text1), len(text2)))
    data.append([text1, text2, dist, round(similarity, 4)])

# Convert to DataFrame
df = pd.DataFrame(data, columns=["Text1", "Text2", "Levenshtein_Distance", "Similarity_Score"])

# Compute the Levenshtein distance matrix for heatmap
n = len(text_pairs)
distance_matrix = np.zeros((n, n))

for i in range(n):
    for j in range(n):
        distance_matrix[i, j] = levenshtein_distance(text_pairs[i][1], text_pairs[j][1])

# Convert to DataFrame for visualization
distance_df = pd.DataFrame(distance_matrix, columns=[f"T{idx+1}" for idx in range(n)], index=[f"T{idx+1}" for idx in range(n)])

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(distance_df, annot=True, cmap="coolwarm", fmt=".0f", linewidths=0.5)
plt.title("Levenshtein Distance Heatmap between Text Pairs")
plt.show()

# Display table
df
```
