# 🐍 Python - Counting Words and Characters in Text 🔢

## ❓ Want to quickly extract valuable info from any text using Python?

## 👉 Solution  
By using the `re` and `collections` libraries, we can easily count word and character frequencies to better understand any text, from tweets to research papers.

## 🔧 How does it work?  
🧹 `re` cleans and splits the text into words, ignoring punctuation and special characters.  
📊 `collections.Counter` then counts how often each word appears.

## 🔎 Why does it matter?  
🌟 Analyzing unstructured text is crucial to discover trends, sentiment, or hidden insights in large data volumes.

# Código Python para Jupyter Notebook (comentarios solo en inglés)

```python
# Import necessary libraries
import re  # for regular expressions to clean and split text
from collections import Counter  # to count word frequency
import matplotlib.pyplot as plt  # for plotting
import nltk  # for sample texts

# Download Shakespeare corpus if not already present
nltk.download('gutenberg')

# Load text from Shakespeare's "Hamlet"
text = nltk.corpus.gutenberg.raw('shakespeare-hamlet.txt')

# Clean and split text into words (lowercase, no punctuation)
words = re.findall(r'\b\w+\b', text.lower())

# Count frequency of each word
word_counts = Counter(words)

# Count total characters (excluding spaces)
char_count = len(re.findall(r'\S', text))

# Print total characters and total words
print(f"Total characters (excluding spaces): {char_count}")
print(f"Total words: {len(words)}")

# Show 10 most common words
common_words = word_counts.most_common(10)
print("Top 10 words:", common_words)

# Prepare data for plotting
words_top = [w for w, c in common_words]
counts_top = [c for w, c in common_words]

# Plot the frequencies of the top 10 words
plt.figure(figsize=(10,6))
bars = plt.bar(words_top, counts_top, color='skyblue')
plt.title('Top 10 Most Common Words in Shakespeare\'s Hamlet')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add frequency labels on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 50, yval, ha='center', va='bottom')

plt.show()


````python
# Import necessary libraries
import re  # for regular expressions to clean and split text
from collections import Counter  # to count word frequency
import matplotlib.pyplot as plt  # for plotting
import nltk  # for sample texts

# Download Shakespeare corpus if not already present
try:
    nltk.data.find('corpora/gutenberg')
except LookupError:
    nltk.download('gutenberg')

# Load text from Shakespeare's "Hamlet"
text = nltk.corpus.gutenberg.raw('shakespeare-hamlet.txt')

# Clean and split text into words (lowercase, no punctuation)
words = re.findall(r'\b\w+\b', text.lower())

# Count frequency of each word
word_counts = Counter(words)

# Count total characters (excluding spaces)
char_count = len(re.findall(r'\S', text))

# Print total characters and total words
print(f"Total characters (excluding spaces): {char_count}")
print(f"Total words: {len(words)}")

# Show 10 most common words
common_words = word_counts.most_common(10)
print("Top 10 words:", common_words)

# Prepare data for plotting
words_top = [w for w, c in common_words]
counts_top = [c for w, c in common_words]

# Plot the frequencies of the top 10 words
plt.figure(figsize=(10,6))
bars = plt.bar(words_top, counts_top, color='skyblue')
plt.title('Top 10 Most Common Words in Shakespeare\'s Hamlet')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add frequency labels on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 50, yval, ha='center', va='bottom')

plt.show()
````
