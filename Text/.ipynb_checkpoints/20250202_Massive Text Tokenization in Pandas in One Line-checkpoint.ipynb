```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# Sample data: customer reviews
data = {"text": [
    "The product is amazing and works great!",
    "Not worth the price, very disappointing experience.",
    "Excellent quality, will buy again!",
    "The worst purchase I've made, very bad quality.",
    "Good value for money, satisfied with my order."
]}

# Create DataFrame
df = pd.DataFrame(data)

# Tokenization in one line
df["tokens"] = df["text"].str.split()

# Flatten token lists and count word frequencies
all_tokens = [word.lower() for tokens in df["tokens"] for word in tokens]
word_counts = Counter(all_tokens)

# Convert to DataFrame for visualization
word_freq_df = pd.DataFrame(word_counts.items(), columns=["Word", "Count"])
word_freq_df = word_freq_df.sort_values(by="Count", ascending=False)

# Plot results
plt.figure(figsize=(8,5))
plt.bar(word_freq_df["Word"].head(10), word_freq_df["Count"].head(10), color='skyblue')
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.title("Top 10 Most Common Words in Customer Reviews")
plt.xticks(rotation=45)
plt.show()
```
