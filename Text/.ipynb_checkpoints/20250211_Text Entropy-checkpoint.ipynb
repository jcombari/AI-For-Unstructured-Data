## üìå **C√≥digo en Python (Jupyter Notebook)**  


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
from scipy.stats import entropy

# üìå Sample text dataset with better entropy variability
texts = [
    "AAAAAAAAAAAAAAAA",  # üî¥ Entropy ~ 0.0 (Single repeated character)
    "ABABABABABABABAB",  # üü† Entropy ~ 1.0 (Two alternating characters)
    "HELLOHELLOHELLOHELLO",  # üü† Entropy ~ 1.3 (Few unique characters)
    "Hello, how are you?",  # üü° Entropy ~ 3.5 (Natural sentence)
    "Python and Pandas are powerful for Data Science!",  # üü° Entropy ~ 3.7 (Technical phrase)
    "1234567890abcdef",  # üü¢ Entropy ~ 4.0 (Alphanumeric sequence)
    "Data, data, data, science, science, science!",  # üü¢ Entropy ~ 3.4 (Repeating structured words)
    "def entropy(text): return sum(-p * log2(p) for p in Counter(text).values())",  # üîµ Entropy ~ 4.3 (Code-like text)
    "¬°Hola! ¬øC√≥mo est√°s? Bienvenido a la comunidad de Data Science.",  # üîµ Entropy ~ 4.2 (Spanish sentence with variation)
    "Lz@#7qT!fD9pC&Xd",  # üî¥ Entropy ~ 4.5 (Completely random characters)
    "üòäüòäüòäüòäüòäüòäüòäüòäüòäüòäüòäüòäüòäüòäüòäüòä",  # üî¥ Entropy ~ 0.0 (Repeated emoji)
    "üë®‚Äçüíªüí°üìäü§ñüìàüî•üìâ‚ö°üåéüîçüéØüíæüé≤üî¢üìëüíª",  # üü¢ Entropy ~ 4.8 (Unique emojis)
]

# üìå Convert texts into a Pandas DataFrame
df = pd.DataFrame({"Text": texts})

# üìå Function to calculate entropy
def calculate_entropy(text):
    """Computes the entropy of a given text using character frequencies."""
    counts = Counter(text)  # Count character frequencies
    probabilities = np.array(list(counts.values())) / sum(counts.values())  # Normalize frequencies
    return entropy(probabilities, base=2)  # Compute entropy

# üìå Apply entropy calculation
df["Entropy"] = df["Text"].apply(calculate_entropy)

# üìå Plot entropy results (Fixed Seaborn warning)
plt.figure(figsize=(10, 6))
sns.barplot(x=df["Entropy"], y=df["Text"], hue=df["Text"], palette="viridis", dodge=False, legend=False)
plt.xlabel("Entropy Value")
plt.ylabel("Text Sample")
plt.title("üî† Text Entropy Analysis")
plt.show()

# üìå Display entropy values
df
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import collections
from math import log2
import matplotlib.font_manager as fm

# üîπ Function to calculate Shannon entropy
def entropy(text):
    counter = collections.Counter(text)  # Count character occurrences
    total_chars = len(text)  # Total number of characters
    probabilities = [count / total_chars for count in counter.values()]  # Compute probabilities
    return -sum(p * log2(p) for p in probabilities)  # Compute entropy formula

# üîπ Sample texts with varying entropy levels
texts = [
    "AAAAAAAAAAAAAAAA",  # Very low entropy (only one character) ‚Üí 0.000
    "ABABABABABABABAB",  # Repeating pattern ‚Üí 1.000
    "Hello, how are you?",  # Natural sentence ‚Üí ~3.5
    "Data Science is amazing!",  # Slightly higher entropy ‚Üí ~3.6
    "1234567890abcdef",  # Random hex string ‚Üí 4.000
    "def entropy(text): return sum(-p * log2(p) for p in probabilities)",  # Python code snippet ‚Üí 4.32
    "Data, data, data, science, science, science!",  # Repeating words but diverse characters ‚Üí ~3.39
    "Lz@#7qT!fD9pC&Xd",  # High randomness ‚Üí 4.000
    "¬°Hola! ¬øC√≥mo est√°s? Bienvenido a la comunidad de ciencia de datos.",  # Spanish text with accents ‚Üí ~4.27
    "ü§ñüí°üìäüîçüöÄüåçüéØüî•üìâüìà"  # Fully emoji-based string ‚Üí High entropy (~4.0)
]

# üîπ Create DataFrame
df = pd.DataFrame({"Text": texts})
df["Entropy"] = df["Text"].apply(entropy)  # Apply entropy function

# üîπ Load a font that supports emojis (Windows: Arial Unicode MS, Linux/Mac: Noto Sans)
font_path = "C:/Windows/Fonts/arial.ttf"  # Adjust the path based on your system
prop = fm.FontProperties(fname=font_path)  # Load font properties
plt.rcParams["font.family"] = prop.get_name()  # Set as default font

# üîπ Plot entropy values using Seaborn
plt.figure(figsize=(10, 5))
sns.barplot(x=df["Entropy"], y=df["Text"], palette="viridis")  # Entropy visualization
plt.xlabel("Entropy Value", fontsize=12)
plt.ylabel("Text Sample", fontsize=12)
plt.title("Entropy Analysis of Different Texts", fontsize=14)
plt.grid(axis="x", linestyle="--", alpha=0.7)

# üîπ Show plot
plt.show()
```
