```python
# Import necessary libraries
import pandas as pd  # For handling data in DataFrame format
import matplotlib.pyplot as plt  # For data visualization
import seaborn as sns  # For better visualization
from langdetect import detect  # Language detection library
import warnings  # To suppress warnings

# Suppress specific warnings to avoid clutter
warnings.filterwarnings("ignore", category=UserWarning, message="Glyph.*missing from font")
warnings.filterwarnings("ignore", category=FutureWarning, message=".*passing `palette`.*")

# Create a sample DataFrame with various greetings in different languages
data = {
    "Text": [
        "Hola, ¬øc√≥mo est√°s?",  # Spanish
        "Hello, how are you?",  # English
        "Bonjour, comment √ßa va?",  # French
        "Hallo, wie geht es dir?",  # German
        "Ciao, come stai?",  # Italian
        "Ol√°, como voc√™ est√°?",  # Portuguese
        "–ü—Ä–∏–≤–µ—Ç, –∫–∞–∫ –¥–µ–ª–∞?",  # Russian
        "‰Ω†Â•ΩÔºå‰Ω†Â•ΩÂêóÔºü",  # Chinese
        "„Åì„Çì„Å´„Å°„ÅØ„ÄÅ„ÅäÂÖÉÊ∞ó„Åß„Åô„ÅãÔºü",  # Japanese
        "Hola, buenos d√≠as",  # Spanish
        "Hi, how are you doing?",  # English
        "Salut, comment va?",  # French
        "Guten Tag, wie geht es?",  # German
        "Salve, come va?",  # Italian
        "Oi, tudo bem?",  # Portuguese
        "–ó–¥—Ä–∞–≤—Å—Ç–≤—É–π, –∫–∞–∫ —Ç—ã?",  # Russian
        "ÊÇ®Â•Ω, ‰Ω†ÊÄé‰πàÊ†∑Ôºü",  # Chinese
        "„Åä„ÅØ„Çà„ÅÜ„Åî„Åñ„ÅÑ„Åæ„Åô„ÄÅ„ÅäÂÖÉÊ∞ó„Åß„Åô„ÅãÔºü",  # Japanese
    ]
}

# Convert dictionary to Pandas DataFrame
df = pd.DataFrame(data)

# Function to detect language using langdetect
def detect_language(text):
    try:
        return detect(text)  # Returns language code (e.g., 'es' for Spanish)
    except:
        return "unknown"  # Handles errors if detection fails

# Apply langdetect to detect languages in the DataFrame
df["LangDetect"] = df["Text"].apply(detect_language)

# Display the DataFrame
print(df)

# Count occurrences of each detected language
lang_counts = df["LangDetect"].value_counts()

# Improved visualization with modern aesthetic
plt.figure(figsize=(12, 7))  # Set figure size
sns.set_style("darkgrid")  # Apply a modern dark grid style
sns.set_palette("muted")  # Set a more subtle color palette

# Create the barplot
ax = sns.barplot(
    x=lang_counts.index, 
    y=lang_counts.values, 
    color='skyblue'  # Choose a soft color
)

# Labels and title with improved styling
plt.xlabel("Detected Language", fontsize=16, fontweight='bold', color='darkblue')  # X-axis label
plt.ylabel("Count", fontsize=16, fontweight='bold', color='darkblue')  # Y-axis label

# Fix to render emoticons properly: Use a font that supports them
plt.title("Language Detection using langdetect", fontsize=20, fontweight='bold', color='darkblue', family='DejaVu Sans')  # Chart title

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, fontsize=14, fontweight='bold', color='darkgreen')
plt.yticks(fontsize=14, fontweight='bold', color='darkgreen')

# Adding a gradient background for enhanced aesthetic appeal
ax.set_facecolor('#f1f1f1')  # Light gray background color for the plot

# Show the plot
plt.show()
```
