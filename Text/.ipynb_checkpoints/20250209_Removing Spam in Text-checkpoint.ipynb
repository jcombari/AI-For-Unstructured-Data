## üî¢ Python Code (Ejecutar en Jupyter Notebook)
```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset with structured and unstructured text
data = {
    "text": [
        "Great product! Highly recommend!",
        "üåü BUY NOW - 50% DISCOUNT! üåü",
        "Customer service was helpful.",
        "Limited time offer! Click here!",
        "Amazing quality, will buy again.",
        "Special PROMO: Get one free!",
    ]
}

# Create a Pandas DataFrame
df = pd.DataFrame(data)

# Define spam keywords
spam_keywords = "buy now|discount|offer|promo|free"

# Detect spam using regex
df["is_spam"] = df["text"].str.contains(spam_keywords, case=False, na=False)

# Filter spam messages
spam_df = df[df["is_spam"]]
clean_df = df[~df["is_spam"]]

# Visualization: Count spam vs. non-spam messages
plt.figure(figsize=(6,4))
df["is_spam"].value_counts().plot(kind="bar", color=["green", "red"], alpha=0.7)
plt.xticks(ticks=[0,1], labels=["Clean", "Spam"], rotation=0)
plt.ylabel("Count")
plt.title("Spam vs. Non-Spam Messages")
plt.show()

# Display the filtered results
print("Spam Messages:")
print(spam_df)
print("\nClean Messages:")
print(clean_df)
```



````python
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset with structured and unstructured text
data = {
    "text": [
        "Great product! Highly recommend!",
        "üåü BUY NOW - 50% DISCOUNT! üåü",
        "Customer service was helpful.",
        "Limited time offer! Click here!",
        "Amazing quality, will buy again.",
        "Special PROMO: Get one free!",
    ]
}

# Create a Pandas DataFrame
df = pd.DataFrame(data)

# Define spam keywords
spam_keywords = "buy now|discount|offer|promo|free"

# Detect spam using regex
df["is_spam"] = df["text"].str.contains(spam_keywords, case=False, na=False)

# Filter spam messages
spam_df = df[df["is_spam"]]
clean_df = df[~df["is_spam"]]

# Visualization: Count spam vs. non-spam messages
plt.figure(figsize=(6,4))
df["is_spam"].value_counts().plot(kind="bar", color=["green", "red"], alpha=0.7)
plt.xticks(ticks=[0,1], labels=["Clean", "Spam"], rotation=0)
plt.ylabel("Count")
plt.title("Spam vs. Non-Spam Messages")
plt.show()

# Display the filtered results
print("Spam Messages:")
print(spam_df)
print("\nClean Messages:")
print(clean_df)
````
