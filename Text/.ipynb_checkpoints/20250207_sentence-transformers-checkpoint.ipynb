```python
# Import necessary libraries
import pandas as pd                # For data manipulation and DataFrame creation
from sentence_transformers import SentenceTransformer  # For generating text embeddings
import matplotlib.pyplot as plt    # For plotting graphs
import numpy as np                 # For numerical operations
from sklearn.decomposition import PCA  # For reducing dimensionality for visualization

# Create a sample DataFrame with text data
data = {
    'id': [1, 2, 3, 4, 5],
    'text': [
        "Artificial Intelligence is transforming industries.",
        "Data Science makes complex data accessible.",
        "Machine learning algorithms power modern solutions.",
        "Deep learning enables breakthrough innovations.",
        "Natural Language Processing unlocks insights from text."
    ]
}
df = pd.DataFrame(data)  # Initialize the DataFrame with sample data

# Load a pre-trained SentenceTransformer model
# The model 'all-MiniLM-L6-v2' is chosen for its balance between speed and accuracy
model = SentenceTransformer('all-MiniLM-L6-v2')

# Compute embeddings for each text entry and store them in a new DataFrame column
df['embeddings'] = df['text'].apply(lambda x: model.encode(x))

# Stack the embeddings into a numpy array for dimensionality reduction
embedding_matrix = np.stack(df['embeddings'].values)

# Apply PCA to reduce the embeddings from their original dimension to 2D for visualization
pca = PCA(n_components=2)
embedding_2d = pca.fit_transform(embedding_matrix)

# Add the 2D coordinates to the DataFrame for plotting
df['x'] = embedding_2d[:, 0]  # First principal component
df['y'] = embedding_2d[:, 1]  # Second principal component

# Plot the 2D embeddings
plt.figure(figsize=(8, 6))  # Set the figure size for clarity
plt.scatter(df['x'], df['y'], color='blue')  # Create a scatter plot with the 2D coordinates

# Annotate each point with its corresponding id for identification
for index, row in df.iterrows():
    plt.annotate(f"{row['id']}", (row['x'], row['y']), textcoords="offset points", xytext=(0, 10), ha='center')

# Set the title and labels for the plot to improve readability
plt.title("2D Visualization of Text Embeddings")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.grid(True)  # Add a grid for easier interpretation of the plot
plt.show()  # Display the plot
```

# üìä Understanding the Graph: Visualizing Text Embeddings  

The graph is a **2D visualization of text embeddings**, helping us see how different pieces of text relate based on their meaning.  

## üîé What does each element represent?  

- **Each point = One text**  
  Each text in our dataset has been converted into a **numerical vector (embedding)** that captures its meaning. Originally, these vectors had many dimensions, but we reduced them to **two dimensions using PCA** to make them plottable.  

- **Position in the graph = Semantic similarity**  
  - Texts with **similar meanings** are positioned **closer** together.  
  - Texts with **different meanings** are positioned **farther apart**.  
  - If two points are close, it suggests their text content shares common themes or context.  

- **X and Y axes = Compressed meaning**  
  - The **X-axis (PCA Component 1)** captures the most significant variation in the text data.  
  - The **Y-axis (PCA Component 2)** captures additional but less dominant variations.  
  - These axes **don‚Äôt have specific meanings** (like time or price in a standard graph), but together they help position texts based on relationships.  

- **Numbers on the graph = Text IDs**  
  Each point is labeled with an **ID number**, so we can identify which text it represents in the dataset.  

## ‚öôÔ∏è Why is this useful?  
By visualizing text embeddings, we can:  
‚úîÔ∏è Understand how texts are grouped based on meaning.  
‚úîÔ∏è Detect patterns or clusters of similar topics.  
‚úîÔ∏è Improve text analysis in machine learning applications.  

In short, this graph **transforms words into numbers and then into a visual representation of meaning**‚Äîallowing us to "see" the relationships between different pieces of text. üöÄ
