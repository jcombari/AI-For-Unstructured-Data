## üìå **C√≥digo en Python (Jupyter Notebook)**  


```python
import pandas as pd
import numpy as np
from math import log2
from collections import Counter
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

# Ignorar las advertencias si es necesario
warnings.filterwarnings("ignore", category=UserWarning)

# Funci√≥n para calcular la entrop√≠a
def entropy(text):
    freq = Counter(text)
    return -sum((freq[char] / len(text)) * log2(freq[char] / len(text)) for char in freq)

# Textos con diversos valores de entrop√≠a
texts = [
    "aaaaaaaaaaaaaaaaaa", 
    "111111111111111",  
    "ZZZZZZZZZZZZZZZZ",  
    "ABABABABABABABAB",  
    "hellohellohello",  
    "123123123",  
    "catcatcat",  
    "ü§ñüí°üìäüîçüöÄüåçüéØüî•üìâüìà" ,  
    "Hello, how are you?",  
    "Data Science is fun!",  
    "Machine Learning is exciting!",  
    "I love data analysis and statistics!",  # Entropy should be around 3.6
    "Lz@#7qT!fD9pC&Xd",  # Entropy should be 4
    "¬°Hola! ¬øC√≥mo est√°s? Bienvenido a la comunidad de ciencia de datos." # Spanish text with accents ‚Üí ~4.27
]

# Calcular las entrop√≠as
entropy_values = [entropy(text) for text in texts]

# Crear el DataFrame
df = pd.DataFrame({
    "Text": texts,
    "Entropy": entropy_values
})

# Cambiar la fuente para que soporte emojis (si est√° disponible en tu sistema)
plt.rcParams['font.family'] = 'Segoe UI Emoji'  # Cambiar a fuente que soporte emojis

# Visualizaci√≥n
plt.figure(figsize=(10, 6))
sns.barplot(x="Entropy", y="Text", hue="Text", data=df, palette="viridis", legend=False)
plt.title("Entropy of Texts")
plt.xlabel("Entropy Value")
plt.ylabel("Text")
plt.show()

# Mostrar los valores de entrop√≠a
df
```
