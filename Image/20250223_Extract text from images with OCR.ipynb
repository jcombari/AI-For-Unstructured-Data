```python
#  Import necessary libraries
# Importing pytesseract for OCR and PIL for image handling
import pytesseract
from PIL import Image
import matplotlib.pyplot as plt

#  Generate a sample image with text using PIL
# This creates an image with simple text that we'll use for OCR extraction
from PIL import ImageDraw, ImageFont

# Create a blank white image
image = Image.new("RGB", (300, 100), color="white")
draw = ImageDraw.Draw(image)

# Add some sample text
# The font may vary depending on your OS; 'arial' should work on Windows
try:
    font = ImageFont.truetype("arial.ttf", 15)
except IOError:
    font = ImageFont.load_default()

draw.text((10, 40), "Hello, Python OCR!", fill="black", font=font)

# Save the image temporarily
image.save("sample_text_image.png")

#  Display the image
# This is useful to visualize the text that will be processed by OCR
plt.imshow(image)
plt.axis("off")  # Hide axis for better visualization
plt.show()

#  Extract text using pytesseract
# pytesseract reads the image and converts any detected text into a string
extracted_text = pytesseract.image_to_string(image)

#  Print the extracted text
# The output should display the exact text from the image
print("Extracted Text:")
print(extracted_text)
```
