# 🐍 Extrae texto de imágenes en segundos con OCR 📸

## ❓ ¿Alguna vez has tenido que copiar texto de una imagen manualmente?
Eso ya es cosa del pasado. Hoy, vamos a automatizarlo con `pytesseract` 🧠.

### 👉 Solución:
Con solo unas líneas de código, puedes extraer texto de imágenes como documentos escaneados o fotos de pizarras. ¡Imagina el tiempo que puedes ahorrar!

### 🔧 ¿Cómo funciona?
`pytesseract` es una herramienta OCR (Reconocimiento Óptico de Caracteres) que convierte imágenes en texto editable. Detecta caracteres en imágenes y los transforma en datos útiles.

### 🔎 ¿Por qué importa?
Automatizar esta tarea es esencial para:
- ✨ Digitalizar documentos físicos sin errores manuales  
- ✨ Analizar datos no estructurados en imágenes  
- ✨ Acelerar flujos de trabajo en sectores como legal, salud y finanzas  

### ✨ Ejemplo real:
Una empresa financiera puede escanear facturas físicas y extraer la información automáticamente para cargarla en su sistema contable, reduciendo errores y costos.

### ⚙️ Impacto en el negocio:
- 💡 Ahorra horas de trabajo manual

```python
# 📌 Import necessary libraries
# Importing pytesseract for OCR and PIL for image handling
import pytesseract
from PIL import Image
import matplotlib.pyplot as plt

# 📌 Generate a sample image with text using PIL
# This creates an image with simple text that we'll use for OCR extraction
from PIL import ImageDraw, ImageFont

# Create a blank white image
image = Image.new("RGB", (300, 100), color="white")
draw = ImageDraw.Draw(image)

# Add some sample text
# The font may vary depending on your OS; 'arial' should work on Windows
try:
    font = ImageFont.truetype("arial.ttf", 15)
except IOError:
    font = ImageFont.load_default()

draw.text((10, 40), "Hello, Python OCR!", fill="black", font=font)

# Save the image temporarily
image.save("sample_text_image.png")

# 📌 Display the image
# This is useful to visualize the text that will be processed by OCR
plt.imshow(image)
plt.axis("off")  # Hide axis for better visualization
plt.show()

# 📌 Extract text using pytesseract
# pytesseract reads the image and converts any detected text into a string
extracted_text = pytesseract.image_to_string(image)

# 📌 Print the extracted text
# The output should display the exact text from the image
print("Extracted Text:")
print(extracted_text)
```
