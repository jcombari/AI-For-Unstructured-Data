# ‚ú® **Python üêç: Find Visually Similar Images with Pandas üîç**

## ‚ùì **Ever wondered how systems recommend content or identify duplicate images?** 
The answer lies in comparing image histograms using Python. Let's dive into how you can use **cv2.compareHist()** to compare images in a DataFrame!

## üëâ **Solution:**  
We‚Äôll use OpenCV to compute the histograms of images and then compare them to find similarities, even with rotated versions.

## üîß **How does it work?**  
To identify visually similar images, we first need to extract the histogram of each image. A histogram represents the distribution of pixel intensities, and images with similar histograms are visually similar. Using **cv2.compareHist()**, we can compare the histograms of two images and get a similarity score. If the histograms are nearly identical, the images are very similar!

## üîé **Why is this important?**  
- It‚Äôs key in **recommendation systems** (think Netflix or YouTube).  
- Detecting duplicate or similar images in datasets (like Google Images or photo storage apps).  
- It aids in **image classification** tasks in machine learning models.

## ‚ú® **Real Example:**  
Imagine you are working with a database of thousands of images, and you need to identify similar ones (e.g., duplicate images). This method is perfect for it! Let‚Äôs see how it works in action.

## ‚öôÔ∏è **Impact in Business:**  
By identifying duplicate or similar images efficiently, businesses can reduce storage costs, enhance user experience in media applications, and improve the quality of AI models.

## üìä **Summary of what we do in the code:**  
- Load sample images from the `skimage` library.  
- Compute histograms for these images.  
- Compare histograms to calculate similarity scores.  
- Visualize the similarity scores using a heatmap and display the images.

```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from skimage import data

# Load sample images (these images are already in grayscale)
image1 = cv2.cvtColor(data.astronaut(), cv2.COLOR_RGB2GRAY)  # Convert to grayscale
image2 = data.camera()  # Already grayscale, no conversion needed
image3 = data.coins()  # Already grayscale, no conversion needed
image4 = data.page()  # Already grayscale, no conversion needed

# Create rotated versions of the images
image3_rotated_90 = cv2.rotate(image3, cv2.ROTATE_90_CLOCKWISE)  # 90 degrees clockwise rotation
image4_rotated_90 = cv2.rotate(image4, cv2.ROTATE_90_CLOCKWISE)  # 90 degrees clockwise rotation
image4_rotated_neg90 = cv2.rotate(image4, cv2.ROTATE_90_COUNTERCLOCKWISE)  # -90 degrees counterclockwise rotation

# Repeat an image to simulate duplication (e.g., using image1 again)
image5 = image1  # Duplicate of image1 to simulate a repeated image

# Add new rotated images
image6 = image3_rotated_90  # Image 6 is the 90 degree rotation of image 3
image7 = image4_rotated_neg90  # Image 7 is the -90 degree rotation of image 4

# Function to compute histograms
def compute_histogram(image):
    """Computes a normalized histogram for a given image."""
    hist = cv2.calcHist([image], [0], None, [256], [0, 256])
    cv2.normalize(hist, hist)
    return hist

# Compute histograms for all images
hist1 = compute_histogram(image1)
hist2 = compute_histogram(image2)
hist3 = compute_histogram(image3)
hist4 = compute_histogram(image4)
hist5 = compute_histogram(image5)
hist3_rotated_90 = compute_histogram(image3_rotated_90)
hist4_rotated_90 = compute_histogram(image4_rotated_90)
hist4_rotated_neg90 = compute_histogram(image4_rotated_neg90)
hist6 = compute_histogram(image6)  # Histogram for image 6 (rotated image 3)
hist7 = compute_histogram(image7)  # Histogram for image 7 (rotated image 4)

# List of images and their rotated counterparts
images = {
    "Image 1": hist1,
    "Image 2": hist2,
    "Image 3": hist3,
    "Image 4": hist4,
    "Image 5": hist5,
    "Image 3 Rotated 90": hist3_rotated_90,
    "Image 4 Rotated 90": hist4_rotated_90,
    "Image 4 Rotated -90": hist4_rotated_neg90,
    "Image 6 (Image 3 Rotated 90)": hist6,
    "Image 7 (Image 4 Rotated -90)": hist7
}

# Initialize an empty dictionary to store similarity scores
similarities = {}

# Loop over the images and compare each pair (avoid duplicates)
for image_name1, hist1 in images.items():
    for image_name2, hist2 in images.items():
        if image_name1 < image_name2:  # Ensure no duplicate comparisons (order matters)
            similarity_score = cv2.compareHist(hist1, hist2, cv2.HISTCMP_CORREL)
            similarities[f"{image_name1} vs {image_name2}"] = similarity_score

# Convert the similarities dictionary into a DataFrame for easier visualization
df = pd.DataFrame([similarities])

# Plot similarity scores using a heatmap
plt.figure(figsize=(12, 8))
ax = sns.heatmap(df, annot=True, cmap="coolwarm", fmt=".2f", annot_kws={"size": 9})
plt.title("Histogram Similarity Scores Between Images (Including Rotated Images)")

# Adjusting labels to avoid overlap
for text in ax.texts:
    if float(text.get_text()) < 0.5:  # Check if value is smaller
        text.set_position((text.get_position()[0], text.get_position()[1] + 0.05))  # Adjust y-position
        text.set_fontsize(8)  # Smaller font size for smaller values

plt.show()

# Display the images for visual comparison
fig, axes = plt.subplots(2, 4, figsize=(15, 10))
axes[0, 0].imshow(image1, cmap='gray')
axes[0, 0].set_title("Image 1 (Astronaut)")
axes[0, 0].axis('off')

axes[0, 1].imshow(image2, cmap='gray')
axes[0, 1].set_title("Image 2 (Camera)")
axes[0, 1].axis('off')

axes[0, 2].imshow(image3, cmap='gray')
axes[0, 2].set_title("Image 3 (Coins)")
axes[0, 2].axis('off')

axes[0, 3].imshow(image4, cmap='gray')
axes[0, 3].set_title("Image 4 (Page)")
axes[0, 3].axis('off')

axes[1, 0].imshow(image5, cmap='gray')
axes[1, 0].set_title("Image 5 (Astronaut Duplicate)")
axes[1, 0].axis('off')

axes[1, 1].imshow(image3_rotated_90, cmap='gray')
axes[1, 1].set_title("Image 3 Rotated 90¬∞")
axes[1, 1].axis('off')

axes[1, 2].imshow(image4_rotated_90, cmap='gray')
axes[1, 2].set_title("Image 4 Rotated 90¬∞")
axes[1, 2].axis('off')

axes[1, 3].imshow(image4_rotated_neg90, cmap='gray')
axes[1, 3].set_title("Image 4 Rotated -90¬∞")
axes[1, 3].axis('off')

plt.tight_layout()
plt.show()
```
