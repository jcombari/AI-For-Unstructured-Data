# üêç Convertir im√°genes en redes neuronales con skimage.feature üñºÔ∏è

‚ùì **¬øSab√≠as que una imagen puede transformarse en una red neuronal?**  
Las im√°genes no son solo p√≠xeles; son estructuras interconectadas llenas de patrones. Hoy exploramos c√≥mo convertir una imagen en un **grafo** utilizando **skimage.feature** y visualizar su estructura como si fuera una red neuronal. üöÄ  

## üëâ **Soluci√≥n: Grafos a partir de im√°genes**
Utilizamos **Canny Edge Detection** para detectar los bordes de una imagen y luego construimos un **grafo**, donde cada nodo representa un punto clave en la imagen.  

üîß **¬øC√≥mo funciona?**  
- üì∏ Tomamos una imagen y la convertimos a escala de grises.  
- üñºÔ∏è Aplicamos **Canny Edge Detection** para identificar los contornos.  
- üîó Construimos un **grafo** donde cada nodo es un p√≠xel de borde y sus conexiones representan relaciones espaciales.  
- üé® Visualizamos la estructura conectada del grafo sobre la imagen original.  

üîé **¬øPor qu√© importa?**  
Los **grafos** son herramientas poderosas en visi√≥n computacional, con aplicaciones en:  
‚úî **Reconocimiento de patrones** üìê - Detecci√≥n de formas y objetos en im√°genes.  
‚úî **Segmentaci√≥n de im√°genes** üé≠ - Identificaci√≥n de regiones en im√°genes m√©dicas y satelitales.  
‚úî **Redes Neuronales en Grafos (GNNs)** üß† - Permiten entrenar modelos para an√°lisis estructural avanzado.  

‚ú® **Ejemplo Real: Grafos en im√°genes m√©dicas**  
En aplicaciones biom√©dicas, los **grafos permiten analizar conexiones neuronales en im√°genes de resonancia magn√©tica**. Tambi√©n se utilizan en la detecci√≥n de c√©lulas cancer√≠genas al modelar sus patrones estructurales.  

‚öôÔ∏è **Impacto en el negocio**  
‚úÖ Optimiza la **segmentaci√≥n y clasificaci√≥n de im√°genes** en visi√≥n computacional.

üìä **Resumen del c√≥digo**  
1Ô∏è‚É£ Carga una imagen de ejemplo con `skimage.data`.  
2Ô∏è‚É£ Aplica **Canny Edge Detection** (`skimage.feature.canny`).  
3Ô∏è‚É£ Construye un **grafo** con `networkx`.  
4Ô∏è‚É£ Visualiza la estructura del grafo superpuesta en la imagen.

```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
from skimage import data, color, feature

# Load a sample image (coins dataset from skimage)
image = data.coins()

# Convert image to grayscale if not already
gray_image = color.rgb2gray(image) if len(image.shape) == 3 else image

# Extract edges using Canny edge detection from skimage.feature
edges = feature.canny(gray_image, sigma=2)

# Create an empty graph
graph = nx.Graph()

# Get the dimensions of the image
height, width = edges.shape

# Add nodes: Each pixel that is an edge becomes a node
for y in range(height):
    for x in range(width):
        if edges[y, x]:  # If it's an edge pixel
            graph.add_node((y, x))

# Add edges: Connect each node to its adjacent edge pixels
for y in range(height - 1):
    for x in range(width - 1):
        if edges[y, x]:
            # Connect horizontally and vertically
            if edges[y + 1, x]:
                graph.add_edge((y, x), (y + 1, x))
            if edges[y, x + 1]:
                graph.add_edge((y, x), (y, x + 1))

# Plot the results
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Original grayscale image
ax[0].imshow(gray_image, cmap="gray")
ax[0].set_title("Original Image")

# Graph representation overlaying edges
ax[1].imshow(gray_image, cmap="gray")
nx.draw(graph, pos={node: node[::-1] for node in graph.nodes()}, node_size=1, edge_color="red", ax=ax[1])
ax[1].set_title("Neural Network Graph Representation (Neural Network Structure)")

plt.show()
```
