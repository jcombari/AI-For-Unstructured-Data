#   Python: Genera un Cdigo QR para Compartir tu WiFi  

 Cuntas veces has tenido que escribir tu contrasea de WiFi para compartirla con amigos o clientes? 

 Con Python puedes generar un QR Code que permita conectar cualquier dispositivo sin escribir nada. Solo escanear y listo!

##  Python Code  
```python

import qrcode  # Import the qrcode library

# WiFi network details
ssid = "YOUR_WIFI_NAME"  # Replace with your WiFi network name (SSID)
password = "YOUR_WIFI_PASSWORD"  # Replace with your WiFi password
encryption = "WPA"  # Encryption type: WPA, WEP, or leave empty for an open network

# WiFi QR code format (recognized by most mobile devices)
wifi_config = f"WIFI:T:{encryption};S:{ssid};P:{password};;"

# Generate the QR code
qr = qrcode.make(wifi_config)

# Save the QR code as an image file
qr.save("wifi_qr.png")

print("QR code generated and saved as 'wifi_qr.png'. Scan it to connect automatically.")

```
