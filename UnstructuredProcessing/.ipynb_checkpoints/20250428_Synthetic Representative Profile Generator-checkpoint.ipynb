#  Python | Synthetic Representative Profile Generator 

##  How do you test your models before deploying them to production?  
Sometimes we forget that a system can fail by using unrepresentative data. What if you could create a realistic version of your users without compromising sensitive data?

##  Solution  
We created a synthetic data generator based on real production patterns.   
Ideal for testing interfaces, recommendation systems, or validating dashboards before production deployment.

##  How does it work?
We define the real-world distributions of users (by age, gender, and preferences) observed in production. Then, we generate 200 synthetic profiles that respect those proportions for testing purposes.

> **Note**: The sample size of 200 profiles was selected under external assumptions as the optimal size for evaluating models in this specific case.

##  Why does it matter?
 Protects privacy  
 Simulates realistic conditions  
 Improves model validation and UX  
 Speeds up development without relying on real data

##  Real-life example or case study:
A streaming platform wants to test its recommender. In production, 40% of users are between 26 and 35 years old, and 25% prefer technology. We create test profiles that reflect this exactly to assess unbiased results.

##  Business impact:
 Safe testing  
 More reliable results  
 Prepared for scaling

##  Summary of what's done in the code:
 We generate 200 random profiles with name, gender, age, and preferred category  
 We respect the real production distributions  
 Visualize preferences as a graph  
 Export profiles to a CSV file for batch loading into the company's core system

#  Python | Generador de Perfiles Sintticos Representativos 

## Cmo pruebas tus modelos antes de llevarlos a produccin?  
A veces olvidamos que un sistema tambin puede fallar por usar datos no representativos. Y si pudieras crear una versin realista de tus usuarios sin comprometer datos sensibles?

Creamos un generador de datos sintticos basado en patrones reales de produccin.   
Ideal para probar interfaces, sistemas de recomendacin, o validar dashboards antes del paso a produccin.

##  Cmo funciona?
Se definen las distribuciones reales de usuarios (por edad, gnero y preferencias) observadas en produccin. Luego, generamos 200 perfiles sintticos que respetan esas proporciones para pruebas.

> **Nota**: El tamao de la muestra de 200 perfiles ha sido seleccionado bajo supuestos externos, como el tamao idneo para evaluar los modelos en este caso especfico.

##  Por qu importa?

```python

import random
import pandas as pd
import matplotlib.pyplot as plt
from faker import Faker
import numpy as np

# Initialize faker for generating fake names
fake = Faker()

# Define real-world distributions from production
age_groups = ["18-25", "26-35", "36-45", "46-60"]
age_weights = [0.2, 0.4, 0.25, 0.15]  # realistic distribution

genders = ["Female", "Male", "Non-binary"]
gender_weights = [0.45, 0.45, 0.10]

categories = ["Technology", "Music", "Sports", "Gaming", "Fashion"]
category_weights = [0.25, 0.2, 0.2, 0.2, 0.15]

# Map age group to actual age range
age_map = {
    "18-25": (18, 25),
    "26-35": (26, 35),
    "36-45": (36, 45),
    "46-60": (46, 60)
}

# Generate synthetic data
profiles = []
for _ in range(200):
    name = fake.name()
    gender = random.choices(genders, weights=gender_weights, k=1)[0]
    age_group = random.choices(age_groups, weights=age_weights, k=1)[0]
    age = random.randint(*age_map[age_group])
    category = random.choices(categories, weights=category_weights, k=1)[0]
    profiles.append({"Name": name, "Gender": gender, "Age": age, "Category": category})

# Convert to DataFrame
df = pd.DataFrame(profiles)

# Visualize category distribution
plt.figure(figsize=(10, 6))
df['Category'].value_counts(normalize=True).plot(kind='bar', color='skyblue')
plt.title('Preference Category Distribution (Synthetic Users)')
plt.ylabel('Proportion')
plt.xlabel('Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
