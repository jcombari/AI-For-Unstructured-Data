# ✨ 🐍 Python: Genera un Código QR para Compartir tu WiFi 📶 ✨

❓ ¿Cuántas veces has tenido que escribir tu contraseña de WiFi para compartirla con amigos o clientes? 🤯

👉 Con Python puedes generar un QR Code que permita conectar cualquier dispositivo sin escribir nada. ¡Solo escanear y listo!

## 🔧 ¿Cómo funciona?

- Usa la librería `qrcode` para generar un código QR.  
- Formatea la información en el estándar `WIFI:T:TYPE;S:SSID;P:PASSWORD;;`.  
- Guarda el código QR como una imagen (`wifi_qr.png`).  

## 🔎 ¿Por qué hacerlo con código y no en una web?
✅ **Seguridad y Privacidad**: No subes tus credenciales a páginas desconocidas.  
✅ **Automatización y Escalabilidad**: Ideal para empresas o eventos.  
✅ **Personalización**: Puedes cambiar el diseño del QR.  
✅ **Funciona sin Internet**: No dependes de una conexión para generarlo.  
✅ **Aprendizaje y Control**: Entiendes cómo se estructuran los QR Codes para redes WiFi.  

## ✨ Ejemplo real:
Un café puede colocar el código QR en la entrada para que los clientes se conecten al WiFi sin preguntar por la clave. 🏪📡  

## ⚙️ Impacto en el negocio:
✔ Mejora la experiencia del usuario.  
✔ Reduce preguntas sobre la contraseña de WiFi.  
✔ Ahorra tiempo y mejora la accesibilidad.  

## 📊 Resumen del código:

1. Define la red WiFi (nombre, clave y tipo de seguridad).  
2. Genera el código QR en base a la información.  
3. Guarda la imagen con el QR listo para ser escaneado.

## 📊 Python Code  
```python

import qrcode  # Import the qrcode library

# WiFi network details
ssid = "YOUR_WIFI_NAME"  # Replace with your WiFi network name (SSID)
password = "YOUR_WIFI_PASSWORD"  # Replace with your WiFi password
encryption = "WPA"  # Encryption type: WPA, WEP, or leave empty for an open network

# WiFi QR code format (recognized by most mobile devices)
wifi_config = f"WIFI:T:{encryption};S:{ssid};P:{password};;"

# Generate the QR code
qr = qrcode.make(wifi_config)

# Save the QR code as an image file
qr.save("wifi_qr.png")

print("QR code generated and saved as 'wifi_qr.png'. Scan it to connect automatically.")

```
