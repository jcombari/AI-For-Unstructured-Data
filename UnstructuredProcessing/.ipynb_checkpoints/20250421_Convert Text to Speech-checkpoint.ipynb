# 🐍 Python: Convertir texto a voz 🗣️ 

❓ **¿Alguna vez te has preguntado cómo transformar texto en audio para hacerlo más accesible?**  
👉 **Hoy vamos a usar Pyttsx3, una librería de Python que convierte texto a voz en tiempo real.**

## 🔧 ¿Cómo funciona?  
Pyttsx3 nos permite generar voz a partir de texto y ajustar aspectos como la velocidad y la tonalidad de la voz, lo que lo hace muy útil para personas con discapacidades visuales o simplemente para hacer que las notas se lean en lugar de tener que leerlas.

## 🔎 ¿Por qué importa?  
Esta técnica no solo mejora la accesibilidad, sino que también puede ser utilizada para optimizar tareas como leer reportes o resultados automáticamente, un factor clave para la productividad en equipos de ciencia de datos o inteligencia artificial.

## ✨ Ejemplo real o Caso práctico:  
En este ejemplo, utilizaremos texto generado por un sistema de análisis de datos para leerlo en voz alta. La aplicación más común es en entornos de trabajo o incluso en aplicaciones educativas, donde la voz se usa para transmitir información de manera eficiente y sin necesidad de mirar una pantalla.

# Code in Jupyter Notebook:

```python
# Importing the pyttsx3 library to convert text to speech
import pyttsx3 

# Initialize the text-to-speech engine
engine = pyttsx3.init()

# Setting properties for voice speed (rate) and volume
engine.setProperty('rate', 150)  # Speed of speech (higher is faster)
engine.setProperty('volume', 1)  # Volume level (from 0.0 to 1.0)

# Getting available voices
voices = engine.getProperty('voices')

# Setting voice (optional, choose a different voice index if available)
engine.setProperty('voice', voices[1].id)  # 0 for male, 1 for female

# Text to be converted into speech
text = "Hello, welcome to the world of Python and Data Science. Let's explore how text to speech works!"

# Make the engine say the text aloud
engine.say(text)

# Wait until the speech is finished
engine.runAndWait()


````python
# Importing the pyttsx3 library to convert text to speech
import pyttsx3 

# Initialize the text-to-speech engine
engine = pyttsx3.init()

# Setting properties for voice speed (rate) and volume
engine.setProperty('rate', 150)  # Speed of speech (higher is faster)
engine.setProperty('volume', 1)  # Volume level (from 0.0 to 1.0)

# Getting available voices
voices = engine.getProperty('voices')

# Setting voice (optional, choose a different voice index if available)
engine.setProperty('voice', voices[1].id)  # 0 for male, 1 for female

# Text to be converted into speech
text = "Hello, welcome to the world of Python and Data Science. Let's explore how text to speech works!"

# Make the engine say the text aloud
engine.say(text)

# Wait until the speech is finished
engine.runAndWait()
````
