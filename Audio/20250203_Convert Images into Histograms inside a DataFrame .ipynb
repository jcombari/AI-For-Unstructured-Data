```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from skimage import data, color

# Load sample images from skimage
image1 = color.rgb2gray(data.astronaut())  # Convert to grayscale
image2 = color.rgb2gray(data.coffee())  # Convert to grayscale

# Define function to compute histogram
def compute_histogram(image, bins=256):
    """Compute histogram of an image and return bin counts."""
    hist, bin_edges = np.histogram(image, bins=bins, range=(0, 1))
    return hist

# Compute histograms for both images
hist1 = compute_histogram(image1)
hist2 = compute_histogram(image2)

# Create DataFrame to store histogram data
df = pd.DataFrame({'Bin': np.arange(256), 'Image1': hist1, 'Image2': hist2})

# Display the images
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Display images in the first subplot
ax[0].imshow(image1, cmap='gray')
ax[0].set_title('Astronaut Image')
ax[0].axis('off')  # Turn off the axis

ax[1].imshow(image2, cmap='gray')
ax[1].set_title('Coffee Image')
ax[1].axis('off')  # Turn off the axis

plt.show()

# Display histograms
plt.figure(figsize=(10, 5))
plt.plot(df['Bin'], df['Image1'], label="Astronaut Image", color='blue')
plt.plot(df['Bin'], df['Image2'], label="Coffee Image", color='red')
plt.xlabel('Pixel Intensity Bins')
plt.ylabel('Frequency')
plt.title('Histogram Comparison of Two Images')
plt.legend()
plt.show()

# Display the first rows of the DataFrame
df.head()
```
