# üéôÔ∏è üêç Python | Analizar emociones en audio con librosa  

## ‚ùì ¬øAlguna vez has querido ense√±arle a una m√°quina a detectar emociones en la voz?  
üòì Con **MFCCs en librosa**, puedes analizar c√≥mo suena la emoci√≥n en el habla y aplicarlo en IA conversacional.  

### üëâ Soluci√≥n  
‚úÖ Extraemos **Coeficientes Cepstrales de Frecuencia Mel (MFCCs)** con `librosa.feature.mfcc`.  
‚úÖ Transformamos audios en datos estructurados.  
‚úÖ Identificamos patrones de emoci√≥n en la voz.  

### üîß ¬øC√≥mo funciona?  
üéôÔ∏è Se captura un audio y se convierte en ondas de sonido.  
üî¨ Extraemos caracter√≠sticas como tono y timbre.  
üìä Luego, analizamos estas caracter√≠sticas con modelos de ML o DL para detectar emociones.  

### üîé ¬øPor qu√© importa?  
üó£Ô∏è Los asistentes virtuales pueden responder mejor si entienden emociones.  
üìû Mejora la experiencia en call centers y atenci√≥n al cliente.  
üß† Puede ayudar en salud mental al analizar tonos emocionales.  

### ‚ú® Ejemplo real  
üó£Ô∏è **Detectar emociones en llamadas de servicio al cliente** para priorizar casos urgentes.  

### ‚öôÔ∏è Impacto en el negocio  
üöÄ Mejora la personalizaci√≥n en atenci√≥n al cliente.  
üéØ Mayor precisi√≥n en asistentes de voz como Alexa o Siri.  
üí° Aplicaciones en entretenimiento, marketing y salud.  

### üìä ¬øQu√© hace el c√≥digo?  
üîπ Carga un audio de muestra.  
üîπ Extrae MFCCs con `librosa`.  
üîπ Visualiza los datos con `matplotlib`.  
üîπ Analiza patrones de emoci√≥n en la voz.  


```python
import librosa  # Library for audio processing
import librosa.display  # For visualizing audio features
import numpy as np  # Numerical operations
import matplotlib.pyplot as plt  # For plotting

# üîπ Load an example audio file (Trumpet sound from Librosa's dataset)
audio_path = librosa.example('trumpet')  # Alternative: Use a local file (e.g., 'your_audio.wav')
y, sr = librosa.load(audio_path, sr=None)  # Load the audio file (y: signal, sr: sample rate)

# üîπ Extract MFCCs (Mel-Frequency Cepstral Coefficients)
mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)  # Extract 13 MFCC features

# üîπ Plot the MFCC features as a spectrogram
plt.figure(figsize=(10, 4))  # Set figure size
librosa.display.specshow(mfccs, x_axis='time', sr=sr)  # Display MFCCs over time
plt.colorbar(label='MFCC Coefficients')  # Add color scale for better visualization
plt.title('MFCC Features of Trumpet Audio')  # Title of the graph
plt.xlabel('Time')  # X-axis label
plt.ylabel('MFCC Coefficients')  # Y-axis label
plt.show()  # Show the plot

# üîπ Print the shape of the extracted MFCCs
print("MFCC Shape:", mfccs.shape)  # The output shape is (13, time_frames), showing 13 MFCCs across time
```
