```python
# Importing necessary libraries
import numpy as np

pixels = image_data.reshape(-1, 3)

scaler = StandardScaler()

````python
# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_sample_image

# Load a sample image (you can replace this with your own image)
china = load_sample_image("china.jpg")
image_data = china / 255.0  # Normalize the image data to 0-1

# Reshaping the image data into a 2D array (pixels x RGB)
pixels = image_data.reshape(-1, 3)

# Standardize the data for KMeans
scaler = StandardScaler()
pixels_scaled = scaler.fit_transform(pixels)

# Apply KMeans clustering to identify dominant colors (we choose 5 clusters for this example)
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(pixels_scaled)

# Get the RGB values of the cluster centers (dominant colors)
dominant_colors = kmeans.cluster_centers_

# Convert the dominant colors back to the original scale (0-1 to 0-255)
dominant_colors = scaler.inverse_transform(dominant_colors)

# Plot the dominant colors
plt.figure(figsize=(8, 6))
plt.imshow([dominant_colors.astype(int)])  # Convert to integer for display
plt.axis('off')
plt.title('Dominant Colors Identified by KMeans')
plt.show()

# Creating a Pandas DataFrame to display the results
dominant_colors_df = pd.DataFrame(dominant_colors.astype(int), columns=['Red', 'Green', 'Blue'])
dominant_colors_df.index = ['Color 1', 'Color 2', 'Color 3', 'Color 4', 'Color 5']

# Show the DataFrame with the dominant colors
dominant_colors_df
````
