#  Extracting a Song's Unique "Fingerprint" with MFCC 

## Ever wondered how apps recognize songs in seconds?
 They use *MFCC* (Mel-Frequency Cepstral Coefficients), a technique that converts audio into a unique "fingerprint."

##  How does it work?
 Every sound has a frequency structure  
 MFCC translates this structure into a set of coefficients  
 These coefficients allow us to compare audios and detect plagiarism.

##  Why does it matter?
 Instant song identification (Shazam-like)  
 Automatic plagiarism detection  
 Genre classification using Machine Learning 

##  Real-world example:
Imagine we want to compare three audios:
 A trumpet sound  
 A completely different sound (water)  
 Another trumpet (same recording as the first one)  

###  Business impact:
 Copyright protection for the music industry  
 Automation in streaming platforms  
 Trend analysis in music genres

#  Extraer la "firma" nica de una cancin con MFCC 

## Alguna vez te has preguntado cmo las apps reconocen canciones en segundos?
 Se basan en *MFCC* (Mel-Frequency Cepstral Coefficients), una tcnica que convierte el audio en una "huella digital" nica.

##  Cmo funciona?
 Cada sonido tiene una estructura frecuencial 
 MFCC traduce esta estructura en un conjunto de coeficientes 
 Con ellos, podemos calcular similitudes entre audios y detectar si una cancin ha sido plagiada.

##  Por qu importa?
 Identificacin rpida de canciones (Shazam-like)  
 Deteccin de plagio automtica  
 Agrupacin de gneros musicales con Machine Learning 

##  Ejemplo real:  
Imagina que queremos comparar tres audios:
 Un sonido de trompeta 
 Un sonido completamente diferente (agua) 
 Otra trompeta (misma grabacin que la primera)

```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
import librosa.display
from scipy.spatial.distance import cosine

# Function to extract MFCC features from an audio array
def extract_mfcc(y, sr, n_mfcc=13):
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
    return np.mean(mfcc, axis=1)  # Averaging across time

# Load built-in audio samples from librosa
y1, sr1 = librosa.load(librosa.example('trumpet'))  # Example sound 1 (Trumpet)
    
y2, sr2 = librosa.load(librosa.example('fishin'))   # Example sound 2 (very different, low similarity)
    
y3, sr3 = librosa.load(librosa.example('trumpet'))    # Example sound 3 (identical to Audio1)

# Extract MFCC signatures
mfcc_1 = extract_mfcc(y1, sr1)
mfcc_2 = extract_mfcc(y2, sr2)
mfcc_3 = extract_mfcc(y3, sr3)

# Compute cosine similarity
dist_1_2 = cosine(mfcc_1, mfcc_2)
dist_1_3 = cosine(mfcc_1, mfcc_3)

# Define a threshold for plagiarism detection (empirical tuning may be needed)
threshold = 0.2

# Print results
if dist_1_2 > threshold:
    print(f"Audio1 is NOT a plagiarism of Audio2 (Similarity: {1 - dist_1_2:.2f})")
else:
    print(f"Audio1 is a plagiarism of Audio2 (Similarity: {1 - dist_1_2:.2f})")

if dist_1_3 > threshold:
    print(f"Audio1 is NOT a plagiarism of Audio3 (Similarity: {1 - dist_1_3:.2f})")
else:
    print(f"Audio1 IS a plagiarism of Audio3 (Similarity: {1 - dist_1_3:.2f})")

# Plot MFCC comparisons
plt.figure(figsize=(8, 4))
plt.bar(["Audio1 vs Audio2", "Audio1 vs Audio3"], [1 - dist_1_2, 1 - dist_1_3], color=['blue', 'red'])
plt.xlabel("Audio Comparison")
plt.ylabel("Similarity Score (1 - Cosine Distance)")
plt.title("Audio Similarity Analysis using MFCC")
plt.ylim(0, 1)
plt.show()

# Plot spectrograms
fig, axes = plt.subplots(3, 1, figsize=(10, 10))
librosa.display.specshow(librosa.amplitude_to_db(np.abs(librosa.stft(y1)), ref=np.max), sr=sr1, x_axis='time', y_axis='log', ax=axes[0])
axes[0].set_title("Spectrogram of Audio1 (Trumpet)")
librosa.display.specshow(librosa.amplitude_to_db(np.abs(librosa.stft(y2)), ref=np.max), sr=sr2, x_axis='time', y_axis='log', ax=axes[1])
axes[1].set_title("Spectrogram of Audio2 (fishin)")
librosa.display.specshow(librosa.amplitude_to_db(np.abs(librosa.stft(y3)), ref=np.max), sr=sr3, x_axis='time', y_axis='log', ax=axes[2])
axes[2].set_title("Spectrogram of Audio3 (Trumpet)")
plt.tight_layout()
plt.show()
```
