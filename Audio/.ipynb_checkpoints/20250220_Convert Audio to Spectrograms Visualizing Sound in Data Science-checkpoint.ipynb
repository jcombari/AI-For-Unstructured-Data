# 🐍 Convertir audio a espectrogramas 🎵📊

## ❓ ¿Sabías que la escalofriante escena de la ducha en *Psycho* (1960) es tan aterradora por su sonido? 🔪🎶  
Las notas agudas y estridentes de los violines en la icónica escena no solo impactan nuestra percepción, sino que pueden visualizarse con espectrogramas 🎨🔊.  

👉 Usaremos `scipy.signal.spectrogram` para convertir un fragmento del audio de *Psycho* en una imagen que revela sus frecuencias más aterradoras.  

## 🔧 ¿Cómo funciona?  
Un espectrograma es una representación visual de cómo la energía de un sonido cambia con el tiempo en diferentes frecuencias. Básicamente, convierte audio en imágenes, lo que lo hace muy útil en Machine Learning para:  

- 🎤 Reconocimiento de voz.  
- 🔊 Análisis de sonidos ambientales.  
- 🎵 Procesamiento de música y efectos de sonido.  

## 🔎 ¿Por qué importa?  
En ciencia de datos aplicada al audio, entender la distribución de frecuencias ayuda a entrenar modelos de ML más precisos 🧠📈. Por ejemplo, en detección de fraudes en llamadas o identificación de emociones en la voz.

```python
import numpy as np
import scipy.signal
import scipy.io.wavfile as wav
import matplotlib.pyplot as plt

# Load the audio file
file_path = "20250220_psycho_audio.wav"
sample_rate, audio_data = wav.read(file_path)

# If the audio is stereo, convert it to mono by averaging both channels
if len(audio_data.shape) > 1:
    audio_data = np.mean(audio_data, axis=1)

# Compute the spectrogram using scipy.signal.spectrogram
frequencies, times, Sxx = scipy.signal.spectrogram(audio_data, fs=sample_rate)

# Add a small value (epsilon) to avoid log(0) issues
eps = 1e-10
Sxx = np.maximum(Sxx, eps)  # Ensure all values are >= eps

# Plot the spectrogram using the 'inferno' colormap
plt.figure(figsize=(10, 6))
plt.pcolormesh(times, frequencies, 10 * np.log10(Sxx), shading='gouraud', cmap='inferno')
plt.colorbar(label="Intensity (dB)")
plt.xlabel("Time (s)")
plt.ylabel("Frequency (Hz)")
plt.title("Spectrogram using scipy.signal.spectrogram")
plt.show()
```
