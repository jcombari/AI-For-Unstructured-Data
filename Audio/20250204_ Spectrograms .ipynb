# üîä Crear espectrogramas de audio en Pandas  

## ‚ùì ¬øSab√≠as que puedes visualizar el sonido como una imagen?  
üôÉ **S√≠! Con Pandas y librosa, transformamos audio en espectrogramas para an√°lisis y machine learning.**

Usaremos `librosa` para convertir audio en espectrogramas y almacenarlos en un DataFrame de `pandas`.

```python
import librosa
import librosa.display
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# üîä Generate a more complex audio signal
sr = 22050  # Sample rate
duration = 3  # seconds
t = np.linspace(0, duration, int(sr * duration), endpoint=False)

# üéµ Create a richer audio wave:
# - Base sine wave at 440 Hz
# - A second harmonic at 880 Hz (double frequency)
# - A third harmonic at 1320 Hz (triple frequency)
# - Frequency modulation to add variation
# - White noise to simulate a real environment

audio = (0.5 * np.sin(2 * np.pi * 440 * t) +   # Base frequency
         0.3 * np.sin(2 * np.pi * 880 * t) +   # Second harmonic
         0.2 * np.sin(2 * np.pi * 1320 * t) +  # Third harmonic
         0.2 * np.sin(2 * np.pi * 220 * t * (1 + 0.5 * np.sin(2 * np.pi * 2 * t))) +  # FM modulation
         0.05 * np.random.randn(len(t)))       # White noise

# üéõÔ∏è Compute spectrogram using Short-Time Fourier Transform (STFT)
n_fft = 2048  # Larger window size for better frequency resolution
hop_length = 512  # Step size for moving the window

stft_result = np.abs(librosa.stft(audio, n_fft=n_fft, hop_length=hop_length, window='hann'))

# üìä Convert spectrogram to a DataFrame
df_spectrogram = pd.DataFrame(stft_result)

# üé® Improved visualization of the spectrogram
plt.figure(figsize=(12, 6))
librosa.display.specshow(librosa.amplitude_to_db(stft_result, ref=np.max),
                         sr=sr, hop_length=hop_length, x_axis="time", y_axis="log", cmap="inferno")
plt.colorbar(format="%+2.0f dB")
plt.title("Spectrogram of a Richer Audio Signal")
plt.xlabel("Time (s)")
plt.ylabel("Frequency (Hz)")

# üñºÔ∏è Save image
plt.savefig("20250204_Spectrograms .png", dpi=300, bbox_inches="tight")  # Save as high-quality PNG

# üì§ Display the saved image
plt.show()

# üîç Display the first few rows of the spectrogram DataFrame
df_spectrogram.head()
```
