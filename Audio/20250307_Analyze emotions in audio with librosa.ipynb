# 🎙️ 🐍 Python | Analizar emociones en audio con librosa  

## ❓ ¿Alguna vez has querido enseñarle a una máquina a detectar emociones en la voz?  
😓 Con **MFCCs en librosa**, puedes analizar cómo suena la emoción en el habla y aplicarlo en IA conversacional.  

### 👉 Solución  
✅ Extraemos **Coeficientes Cepstrales de Frecuencia Mel (MFCCs)** con `librosa.feature.mfcc`.  
✅ Transformamos audios en datos estructurados.  
✅ Identificamos patrones de emoción en la voz.  

### 🔧 ¿Cómo funciona?  
🎙️ Se captura un audio y se convierte en ondas de sonido.  
🔬 Extraemos características como tono y timbre.  
📊 Luego, analizamos estas características con modelos de ML o DL para detectar emociones.  

### 🔎 ¿Por qué importa?  
🗣️ Los asistentes virtuales pueden responder mejor si entienden emociones.  
📞 Mejora la experiencia en call centers y atención al cliente.  
🧠 Puede ayudar en salud mental al analizar tonos emocionales.  

### ✨ Ejemplo real  
🗣️ **Detectar emociones en llamadas de servicio al cliente** para priorizar casos urgentes.  

### ⚙️ Impacto en el negocio  
🚀 Mejora la personalización en atención al cliente.  
🎯 Mayor precisión en asistentes de voz como Alexa o Siri.  
💡 Aplicaciones en entretenimiento, marketing y salud.  

### 📊 ¿Qué hace el código?  
🔹 Carga un audio de muestra.  
🔹 Extrae MFCCs con `librosa`.  
🔹 Visualiza los datos con `matplotlib`.  
🔹 Analiza patrones de emoción en la voz.  




```python
import librosa  # Library for audio processing
import librosa.display  # For visualizing audio features
import numpy as np  # Numerical operations
import matplotlib.pyplot as plt  # For plotting

# 🔹 Load an example audio file (Trumpet sound from Librosa's dataset)
audio_path = librosa.example('trumpet')  # Alternative: Use a local file (e.g., 'your_audio.wav')
y, sr = librosa.load(audio_path, sr=None)  # Load the audio file (y: signal, sr: sample rate)

# 🔹 Extract MFCCs (Mel-Frequency Cepstral Coefficients)
mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)  # Extract 13 MFCC features

# 🔹 Plot the MFCC features as a spectrogram
plt.figure(figsize=(10, 4))  # Set figure size
librosa.display.specshow(mfccs, x_axis='time', sr=sr)  # Display MFCCs over time
plt.colorbar(label='MFCC Coefficients')  # Add color scale for better visualization
plt.title('MFCC Features of Trumpet Audio')  # Title of the graph
plt.xlabel('Time')  # X-axis label
plt.ylabel('MFCC Coefficients')  # Y-axis label
plt.show()  # Show the plot

# 🔹 Print the shape of the extracted MFCCs
print("MFCC Shape:", mfccs.shape)  # The output shape is (13, time_frames), showing 13 MFCCs across time
```
