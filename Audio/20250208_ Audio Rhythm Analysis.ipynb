```python
import librosa
import librosa.display
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load a sample audio file (Librosa provides built-in samples)
audio_path = librosa.example('trumpet')  # Sample trumpet audio

# Load audio file and sample rate
y, sr = librosa.load(audio_path, sr=None)

# Extract tempo (BPM)
tempo, _ = librosa.beat.beat_track(y=y, sr=sr)

# Extract pitch (spectral centroid)
spectral_centroid = librosa.feature.spectral_centroid(y=y, sr=sr)
mean_pitch = np.mean(spectral_centroid)

# Create a Pandas DataFrame with extracted features
df = pd.DataFrame({
    "Feature": ["BPM", "Mean Pitch"],
    "Value": [tempo, mean_pitch]
})

# Print the DataFrame
print(df)

# Visualization of spectral centroid
plt.figure(figsize=(10, 4))
plt.semilogy(spectral_centroid.T, label='Spectral Centroid', color='b')
plt.ylabel('Frequency (Hz)')
plt.xlabel('Time')
plt.title('Spectral Centroid Over Time')
plt.legend()
plt.show()
```
